```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("data.csv")
df.head()
```

```python
df = df.rename(columns = {"diagramA_x": "wantedX", "diagramA_y": "wantedY", "diagramB_x": "gotX", "diagramB_y": "gotY"})
df = df.drop(columns = {"ip", "user_agent", "referrer"})
df.head()
```

```python
def to_pixels(v):
    if (0 <= v <= 1.000001):
        return v * 600
    else:
        return v
    
for col in ["wantedX", "wantedY", "gotX", "gotY"]:
    df[col + "_px"] = df[col].apply(to_pixels)

df.head()
```

```python
# Horrific circle maths
RADIUS = 114.12
INTERDISCIPLINARY_X, INTERDISCIPLINARY_Y = 249.12, 253.12
PROBLEMS_X, PROBLEMS_Y = 363.12, 253.12
METHODS_X, METHODS_Y = 307.12, 367.12

# These are the actual circles
CIRCLES = {
    "I": (INTERDISCIPLINARY_X, INTERDISCIPLINARY_Y),
    "P": (PROBLEMS_X, PROBLEMS_Y),
    "M": (METHODS_X, METHODS_Y),
}

def inside_circle(x, y, cx, cy, r=RADIUS):
    return (x - cx)**2 + (y - cy)**2 <= r**2

def classify_point(x, y):
    hits = []

    for key, (cx, cy) in CIRCLES.items():
        if inside_circle(x, y, cx, cy):
            hits.append(key)

    # e.g. ["I", "P"] → "IP", ["I","M","P"] → "IMP"
    code = "".join(sorted(hits))

    return code
```

```python
df["wanted_region"] = df.apply(
    lambda row: classify_point(row["wantedX_px"], row["wantedY_px"]),
    axis = 1
)

df["got_region"] = df.apply(
    lambda row: classify_point(row["gotX_px"], row["gotY_px"]),
    axis = 1
)

df.head()
```

```python
IDx = INTERDISCIPLINARY_X - RADIUS
IDy = INTERDISCIPLINARY_Y - np.tan(2.0944) * RADIUS # 120 degrees in Radians
ProbX = PROBLEMS_X + RADIUS
ProbY = PROBLEMS_Y - np.tan(2.0944) * RADIUS
MethodsX = METHODS_X
MethodsY = METHODS_Y + RADIUS

def distance(x1, y1, x2, y2):
    return np.sqrt((x1-x2)**2 + (y1-y2)**2)

def calculate_distance(row):
    return distance(row["wantedX_px"], row["wantedY_px"], row["gotX_px"], row["gotY_px"])

def calculate_strength(row, coordtype="got"):
    row[f"{coordtype}_ID_strength"] = distance(row[f"{coordtype}X_px"], row[f"{coordtype}Y_px"], IDx, IDy)
    row[f"{coordtype}_Prob_strength"] = distance(row[f"{coordtype}X_px"], row[f"{coordtype}Y_px"], ProbX, ProbY)
    row[f"{coordtype}_Methods_strength"] = distance(row[f"{coordtype}X_px"], row[f"{coordtype}Y_px"], MethodsX, MethodsY)
    return row
```

```python
bruno = df[df["email"] == "bruno.collingridge@lis.ac.uk"]
bruno
```

```python
calculate_strength(bruno)
calculate_strength(bruno, coordtype="wanted")
```

```python
def calculate_strength_change(row):
    ID_change = row["got_ID_strength"] - row["wanted_ID_strength"]
    Prob_change = row["got_Prob_strength"] - row["wanted_Prob_strength"]
    Methods_change = row["got_Methods_strength"] - row["wanted_Methods_strength"]
    return ID_change, Prob_change, Methods_change

calculate_strength_change(bruno)
```
